## What is Python?

Python is a simple yet powerful programming language. 

Some features of this language:

- **Free**: A lot of the resources for learning the language are free, too.
- **Interpreted**: This means the source code (the programs you write) are executed as-is without first compiling into an executable (e.g. an `.exe` program on Windows).
- **Cross-Platform**: Code you write on your Mac will work equally well on Linux and Windows (with a few exceptions that we won't worry about). You can also use online "Python Notebooks" -- like the page you're reading now!

## Why use Python for Machine Learning?

Python is one of several programming languages you can use for machine learning. 

Several other languages are **specifically** designed for numerical analysis:

- Julia
- R
- SASS
- Statistica
- MATLAB

Python is actually a "general purpose" language, like C++, Java, or Rust, meaning that it can be used for all kinds of programming, not just numerical analysis. However, there are a few things that Python has that make it ideal for learning about machine learning and artificial intelligence.

- **Popular:**  90% of the time, you can quickly Google answers to any problems or questions you have, and the vast majority of machine learning tutorials are based on Python. 
- **Mature:** The core language of Python has been "augmented" with packages which you can import into your code to do complex things like plotting graphics, training a model, and interacting with files.
- **Easy to Learn:** In other langauges, you need to know things about computer architecture, low-level memory management, and other details. Doing something simple, like creating and saving a plot, can be very complex to do from scratch. Python's package system makes these things easy to do.

## Python Terminology

So how do you "do Python"? There are several ways you can write and execute Python code; below are the three main ones for developers.


### Interactive Interpreter (REPL)





On the command line (Terminal in MacOS, or Command Prompt in Windows) you can type `python` and hit enter. This will start the interpreter, also known as the **REPL** (Read-Eval-Print Loop): It will **Read** what you type on a line, **Evaluate** the line as a bit of Python code, and **Print** the results of the evaluation. It will do this over and over in a **Loop**.

This isn't very efficient for longer bits of code, but if you just want to check something real quick, this works.

### Scripts and Function

In a text editor, you can write files that end in a `.py` extension. These files are basically a long list of Python commands, i.e. "source code". If you write a script called `runme.py`, you can execute it from the command line by typing `python runme.py`. 

Your script can contain statements / expressions, functions, import external packages, and even reference other `.py` files in the file system: You could have a `main.py` function which calls functions defined in a `utils.py` file. To distribute your code, you must give someone all the local `.py` files so they can run your functions. 

### Local Notebooks

A "notebook" is basically a file that you can view in a web browser. When you open a notebook, you typically see a set of text fields that can be either "code cells" or "text cells". When you type Python code into the code cell and hit `shift` + `enter`, the code is sent to a Python program running in the background, which interprets the code and spits back the result. The whole thing is running in a single "session", so you can keep executing more code blocks and it will remember what you executed previously.

```python
a = 6
b = 2
c = a + b
print(c)
```

```python
d = c ** 2
```

```python
print(d)
```

In my opinion, for learning and demonstration purposes, notebooks are the best. You can test code one chunk at a time, looking at the output as you go to make sure it's doing what you think it's doing. The fact that it's non-linear is great for making quick modifications.

## Deep Learning and Artificial Intelligence



In Python, there are two major libraries that most folks use to build, train, and evaluate deep learning models: [PyTorch](https://pytorch.org), which is developed and sponsored by Facebook, and [TensorFlow](https://tensorflow.org), which is developed and sponsored by Google.

As you can imagine, there are a lot of opinions on the Internet about which package to use and which is "better." Unless you're doing deep learning research, it doesn't matter which one you use: most basic algorithms are implemented in both, and new algorithms published in one framework are often ported to the other by the community fairly quickly.

### PyTorch
